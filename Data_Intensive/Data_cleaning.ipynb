```python
import pandas as pd
accidents = pd.read_csv("Acc.csv")
```

```python
shoes = pd.read_csv("7004_1.csv",error_bad_lines = False)
shoes.columns
```

```python
#Changing the date-format in the shoes dataset to match with that in accidents 
shoes.dateAdded.tolist()
shoes['Date'] = shoes.dateAdded.str.split('T').str[0]
shoes.head()
shoes['Date'] = pd.to_datetime(shoes['Date'])
shoes['Date'] = shoes['Date'].dt.strftime('%d/%m/%Y')
shoes.head()
```

```python
#taking only relevant columns
shoe_f = shoes.filter(['Date', 'colors','prices.amountMin'], axis = 1)
```

```python
#renaming a column
shoe_f = shoe_f.rename(columns = {'prices.amountMin':'ShoesPrices'})
shoe_f.head()
```

```python
shoe_f = shoe_f.dropna()
```

```python
shoe_f.head()
```

```python
#Adding a column to indicate whether the color of shoes is black or not 
shoe_f['colors'] = shoe_f['colors'].str.lower() 
shoe_f['Black_shoes'] = shoe_f['colors'].apply(lambda x: sum(i in r".black" for i in x.split()))
shoe_f.head()
```

```python
#cleaning values
shoe_f['Black_shoes'].loc[shoe_f['Black_shoes'] > 1] = 1
shoe_f.head()
```

```python
accidents.head()
```

```python
#Taking relevant columns & Renaming
accidents_f = accidents.filter(['Date', 'Number_of_Vehicles','Did_Police_Officer_Attend_Scene_of_Accident'], axis=1)
accidents_f = accidents_f.rename(columns = {'Did_Police_Officer_Attend_Scene_of_Accident':'Investigation'})
accidents_f.head()
```

```python
#cleaning values
accidents_f['Investigation'].loc[accidents_f['Investigation'] > 2 ] = 2
accidents_f.head()
```

```python
#Making sure the accidents dataset has the same date format 
accidents_f.Date.tolist()
accidents_f['Date'] = accidents_f.Date.str.split('T').str[0]
#accidents_f.head()
accidents_f['Date'] = pd.to_datetime(accidents_f['Date'])
accidents_f['Date'] = accidents_f['Date'].dt.strftime('%d/%m/%Y')
accidents_f.head()
```

```python
shoe_f.to_csv(r'shoes_f.csv',header = False, index = False)
accidents_f.to_csv(r'accidents_f.csv',header = False, index = False)
```
